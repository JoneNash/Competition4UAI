```python
#读取数据集
father_path="/Users/leidelong/competition/uai/UAI_Data/"
train_data=pd.read_csv(father_path+"training_data.csv")

```

```python
print train_data.columns
```

```python
target='count'
IDcol = 'index'
startID='start_geo_id'
endID='end_geo_id'
date='create_date'
hour='create_hour'
predictors = [x for x in train_data.columns if x not in [target, IDcol,startID,endID,date,hour]]

```

```python
train_data.head(10)
```

```python
print ('空值占比：')
line_count=train_data[target].count()
for i in predictors :
    nan_cnt=  line_count - train_data[i][train_data[i] >=0].count()
    if nan_cnt!=0:
        print (i,": {:.5f}% ".format(nan_cnt/(line_count*1.0)*100 ))
```
